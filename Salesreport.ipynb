```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Sales Data.csv")
data_cleaned = df.drop(columns=['Unnamed: 0'])
data_cleaned.isnull().sum()
data_cleaned.dropna(inplace=True)
data_cleaned.isnull().sum()
```

```python
import seaborn as sns
monthly_sales = data_cleaned.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='blue')
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=15)
plt.xticks(range(1, 13))
plt.grid(visible=True, linestyle='--', alpha=1)
plt.show()
```

```python
product_sales = data_cleaned.groupby('Product')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(15, 8))
plt.bar(product_sales.index, product_sales.values, color=["red","skyblue","lightgreen","orange"])
plt.title('Product-wise Sales Distribution', fontsize=14)
plt.xlabel('Product Names', fontsize=12)
plt.ylabel('Total Sales-In Dollalrs($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
product_sales
```

```python
data_cleaned=data_cleaned.drop(columns=[])
data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
hourly_sales = data_cleaned.groupby('Hour')['Sales'].sum()
plt.figure(figsize=(8, 5))
plt.bar(hourly_sales.index, hourly_sales.values, color='coral')
plt.title('Hourly Sales Analysis', fontsize=14)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(range(24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
